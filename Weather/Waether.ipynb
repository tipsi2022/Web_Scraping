```python
import re
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
base_link = "https://www.weather-forecast.com/locations/"
```

```python
city = input('Enter City Name --> ')
```

```python
def data_days(soup):
    days_data = soup.find_all('div', class_='b-forecast__table-days-name')

    days = []
    for i in range(12):
        days.append(days_data[i].get_text())
    
    return days
```

```python
def data_date(soup):
    date_data = soup.find_all('div', class_='b-forecast__table-days-date')

    date = []
    for i in range(12):
        date.append(date_data[i].get_text())
    
    return date
```

```python
def data_high(soup):
    high_table = soup.find('tr', class_='b-forecast__table-max-temperature')
    high_data = high_table.find_all('span', class_='temp b-forecast__table-value')

    high_am = []
    high_pm = []
    high_night = [] 
    for i in range(0, len(high_data), 3):
        high_am.append(high_data[i].get_text())
        high_pm.append(high_data[i+1].get_text())
        high_night.append(high_data[i+2].get_text())
    
    return high_am, high_pm, high_night
```

```python
def data_low(soup):
    low_table = soup.find('tr', class_='b-forecast__table-min-temperature')
    low_data = low_table.find_all('span', class_='temp b-forecast__table-value')

    low_am = []
    low_pm = []
    low_night = [] 

    for i in range(0, len(low_data), 3):
        low_am.append(low_data[i].get_text())
        low_pm.append(low_data[i+1].get_text())
        low_night.append(low_data[i+2].get_text())
    
    return low_am, low_pm, low_night
```

```python
def data_humidity(soup):
    hum_table = soup.find('tr', class_='b-forecast__table-humidity')
    hum_data = hum_table.find_all('span', class_='b-forecast__table-value')

    humidity_am = []
    humidity_pm = []
    humidity_night = [] 
    for i in range(0, len(hum_data), 3):
        humidity_am.append(hum_data[i].get_text())
        humidity_pm.append(hum_data[i+1].get_text())
        humidity_night.append(hum_data[i+2].get_text())
    
    return humidity_am, humidity_pm, humidity_night
```

```python
def extract_data(base_link, city):
    link = f"{base_link}{city}/forecasts/latest"
    page = requests.get(link)
    soup = bs(page.content, 'html.parser')
    
    days = data_days(soup)
    date = data_date(soup)
    high_am, high_pm, high_night = data_high(soup)
    low_am, low_pm, low_night = data_low(soup)
    humidity_am, humidity_pm, humidity_night = data_humidity(soup)
    
    return days, date, high_am, high_pm, high_night, low_am, low_pm, low_night, humidity_am, humidity_pm, humidity_night 
```

```python
days, date, high_am, high_pm, high_night, low_am, low_pm, low_night, humidity_am, humidity_pm, humidity_night = extract_data(base_link, city)
```

```python
df = pd.DataFrame()

df['Day'] = days
df['Date'] = date
df['Max Temperature AM'] = high_am
df['Max Temperature PM'] = high_pm
df['Max Temperature Night'] = high_night
df['Min Temperature AM'] = low_am
df['Min Temperature PM'] = low_pm
df['Min Temperature Night'] = low_night
df['Humidity AM'] = humidity_am
df['Humidity PM'] = humidity_pm
df['Humidity Night'] = humidity_night

df
```

```python
df.to_csv('Weather Data.csv', index=False)
```
